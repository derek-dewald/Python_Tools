```python

def materialize_ranges(df, rng, suffix_low="_low", suffix_high="_high",
                       decimals=2, discrete_cols=None, clip_nonnegative=True):
    """
    For every pair of columns like X_low / X_high, create X by sampling uniformly between them.
    """
    df = df.copy()
    discrete_cols = set(discrete_cols or [])

    range_bases = sorted({
        c[:-len(suffix_low)]
        for c in df.columns
        if c.endswith(suffix_low) and (c[:-len(suffix_low)] + suffix_high) in df.columns
    })

    for base in range_bases:
        low = df[f"{base}{suffix_low}"].to_numpy(dtype=float)
        high = df[f"{base}{suffix_high}"].to_numpy(dtype=float)

        # Safety: swap if any bad rows
        swap = high < low
        if swap.any():
            low2 = low.copy()
            low[swap] = high[swap]
            high[swap] = low2[swap]

        u = rng.random(len(df))
        vals = low + (high - low) * u

        if base in discrete_cols:
            vals = np.floor(vals).astype(int)
        else:
            vals = np.round(vals, decimals)

        if clip_nonnegative:
            vals = np.clip(vals, 0, None)

        df[base] = vals

    return df

def create_branches(
    unique_records=40,
    legacy_distribution=legacy_distribution,
    total_mbrs=160000,
    skew=1.25,
    rng=None):

    '''
    
    
    '''
    # If rng not defined, define it.
    rng = np.random.default_rng() if rng is None else rng
    
    # 1) create Branch DataFrame
    branch_df = pd.DataFrame([f"BRANCHNAME {i+1}" for i in range(unique_records)],columns=['BRANCHNAME'])

    # 2) legacy assignment
    branch_df["LEGACY"]  = rng.choice([f"LEGACY_{x+1}" for x in range(len(legacy_distribution))], size=unique_records, p=np.array(legacy_distribution))

    # Assign City by first Identifying which City Dict is to apply, then randomly Selecting
    branch_df["CITY"] = branch_df["LEGACY"].apply(
    lambda legacy: rng.choice(
        list(legacy_comp_dict[legacy].keys()),
        p=np.array(list(legacy_comp_dict[legacy].values())) /
          np.sum(list(legacy_comp_dict[legacy].values())))
    )
        
    # 4) multipliers/factors
    branch_df["MTG_MULTIPLIER"] = branch_df["LEGACY"].map(general_assumptions["MTG_Multiplier"]["value"])
    branch_df["BEF"] = rng.uniform(0.95, 1.09, size=unique_records)
    branch_df["LEF"] = branch_df["LEGACY"].map(general_assumptions["Legacy_efficieny_factor_dict"]["value"])
    branch_df["EXP_GROWTH"] = branch_df["LEF"] * branch_df["BEF"]


    # Select the Default Member Composition Profile to which Classification members will be utilized
    branch_df['BRANCH_MBR_COMPOSITION_CLASS'] = [rng.choice(list(branch_mbr_comp_dict.keys()),p=[x['perc_'] for x in branch_mbr_comp_dict.values()]) for x in range(0,len(branch_df))]

    # Total Number of Members based on Perc allocated and total number as defined in General Assumptions
    branch_df['MEMBERS'] = [int(np.round(x*total_mbrs,0)) for x in rng.dirichlet(np.ones(unique_records))]

    return branch_df

def generate_binary_flag_from_probability(df, col, rng, out_col=None):
    df = df.copy()
    out_col = out_col or col
    p = df[col].to_numpy(dtype=float)
    p = np.clip(p, 0, 1)
    df[out_col] = (rng.random(len(df)) < p).astype(int)
    return df
```

```python
def create_members(branch_df, 
                   mbr_profile_df=None,
                   growth_profile_df=None,
                   branch_mbr_comp_dict=None, 
                   start_date=datetime.datetime(2024,1,1), 
                   rng=None):

    if not rng:
        rng = np.random.default_rng() if rng is None else rng
        
    blocks = []
    next_id = 1

    for row in branch_df.itertuples(index=False):
        n = int(row.MEMBERS)
        if n <= 0:
            continue

        # --- Stage 1: skeleton
        member_ids = np.arange(next_id, next_id + n, dtype=int)
        next_id += n

        tmp = pd.DataFrame({
            "MEMBERNBR": member_ids,
            "STATUS": "Active",
            "MONTH": start_date,
            "BRANCHNAME": row.BRANCHNAME,
            "CITY": row.CITY,
            "LEGACY": row.LEGACY,
            "EXP_GROWTH": row.EXP_GROWTH,
            "BRANCH_MBR_COMPOSITION_CLASS": row.BRANCH_MBR_COMPOSITION_CLASS,
        })

        # --- Stage 2: assign member archetype within branch based on that branch template
        template = branch_mbr_comp_dict[row.BRANCH_MBR_COMPOSITION_CLASS]["profile"]
        classes = np.array(list(template.keys()))
        p = np.array(list(template.values()), dtype=float)
        p = p / p.sum()

        tmp["CLASSIFICATION"] = rng.choice(classes, size=n, p=p)

        # merge profile ranges (low/high columns) so you can sample next
        tmp = tmp.merge(mbr_profile_df, on="CLASSIFICATION", how="left")

        blocks.append(tmp)

    mbr_df = pd.concat(blocks, ignore_index=True)

    mbr_df = materialize_ranges(mbr_df,rng)

    mbr_df = generate_binary_flag_from_probability(mbr_df,'primary_is_beem',rng)
    mbr_df = generate_binary_flag_from_probability(mbr_df,'home_owner',rng)
    mbr_df = generate_binary_flag_from_probability(mbr_df,'kids',rng)
    mbr_df['mortgage_debt'] = mbr_df['mortgage_debt']* mbr_df['home_owner']

    mbr_df['perc_spending_known'] = mbr_df['bill_pay_debit'] +mbr_df['pos_txn_debit'] + mbr_df['investment_debit']
    mbr_df['perc_income_known'] = mbr_df['other_deposit'] +mbr_df['payroll_deposit']

    for tran_perc in ['pos_txn_debit','bill_pay_debit','payroll_deposit','other_deposit']:
        mbr_df[tran_perc] = np.round(mbr_df[tran_perc]*mbr_df['annual_income']*(1/12),2)

    mbr_df['mortgage_bal'] = np.round(mbr_df['mortgage_debt']*mbr_df['primary_is_beem'],2)
    mbr_df['deposit_bal']  = np.round(rng.uniform(0.5, 1.0, size=len(mbr_df))* mbr_df['liquid_assets']*mbr_df['primary_is_beem'],2)

    mbr_df['MBR_GROWTH_PROFILE'] = rng.choice(list(mbr_growth_attrition_profiles.keys()),size=len(mbr_df))

    mbr_df = mbr_df.merge(growth_profile_df,on='MBR_GROWTH_PROFILE',how='left')
    return mbr_df

    cols = [x for x in mbr_df if (x.find('_low')==-1)&(x.find('_high')==-1)]

    return mbr_df[cols]
```
